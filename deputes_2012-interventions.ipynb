```python
# final:
# id_intervention_1, Accoyer , Morlier
# id_intervention_1, Accoyer , Gayou

list_parlementaires = table_parlementaire_orig['nom_de_famille'].values

def list_cited_guys(x):
    y = [];
    for p in list_parlementaires:
        if(x.find(p) >= 0):
            y.append(p)
    return "|".join(y);


ADS_Filtered1 = ADS_Filtered.copy()
ADS_Filtered1['list_of_cited'] = ADS_Filtered1['intervention'].apply(lambda x  : list_cited_guys(x))
ADS_Filtered1.head()
```
